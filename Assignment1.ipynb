## 2
What is the cumulative distribution function (CDF) that corresponds to the following probability distribution function? Please state the value of the CDF for all possible values of $x$.

$f(x) = \begin{cases}
    \frac{1}{3} & 0 < x < 3 \\
    0           & \text{otherwise}
    \end{cases}$

**ANSWER**  
The cumulative distribution function F(x) = $\int_{-\infty}^x  f(x) \, \mathrm{d}x$  
When $x < 0$, $F(x) = 0$;  
When $0 \leq x \leq 3$, $ F(x) = \int_{0}^x  \tfrac{1}{3} \, \mathrm{d}x = \tfrac{1}{3} x $   
When $3 < x $, $F(x) = \int_{0}^{3}  \tfrac{1}{3} \, \mathrm{d}x = 1 $   
Hence,  
$F(x) = \begin{cases}
                0           & x < 0  \\
                \tfrac{1}{3} x  & 0 \leq x \leq 3 \\
                1          & 3 < x
            \end{cases}$

## 7
Speed comparison between vectorized and non-vectorized code. Begin by creating an array of 10 million random numbers using the numpy random.randn module. Compute the sum of the squares first in a for loop, then using Numpy's `dot` module. Time how long it takes to compute each and report the results and report the output. How many times faster is the vectorized code than the for loop approach?

*Note: all code should be well commented, properly formatted, and your answers should be output using the `print()` function as follows (where the # represents your answers, to a reasonable precision):

`Time [sec] (non-vectorized): ######`

`Time [sec] (vectorized):     ######`

`The vectorized code is ##### times faster than the vectorized code`

**ANSWER**

```python
import numpy as np
import time

# Generate the random samples
random_sam=np.random.randn(1,10000000)

# Compute the sum of squares the non-vectorized way (using a for loop)
start_non_vec=time.time();
sum_non_vec=0
for i in range(0,10000000):
    sum_non_vec+=random_sam[0,i]**2
end_non_vec=time.time();
time_non_vec=end_non_vec-start_non_vec; #calculate the running time

# Compute the sum of squares the vectorized way (using numpy)
start_vec=time.time();
sum_vec=random_sam.dot(random_sam.transpose())
end_vec=time.time();
time_vec=end_vec-start_vec;   #calculate the running time

# Print the results
print('Time [sec] (non-vectorized): ', round(time_non_vec,4))
print('Time [sec] (vectorized):     ', round(time_vec,4))
print('The vectorized code is {} times faster than the non-vectorized code'.format(round(time_non_vec/time_vec,1)))
```

The function below computes the function $f(x,y) = x^2 - 2 y^2$ and determines whether this quantity is above or below a given threshold, `thresh=0`. This is done for $x,y \in \{-4,4\}$, over a 2,000-by-2,000 grid covering that domain.

(a) Vectorize this code and demonstrate (as in the last exercise) the speed increase through vectorization and (b) plot the resulting data - both the function $f(x,y)$ and the thresholded output - using [`imshow`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html?highlight=matplotlib%20pyplot%20imshow#matplotlib.pyplot.imshow) from `matplotlib`.

*Hint: look at the `numpy` [`meshgrid`](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.meshgrid.html) documentation*

```python
import numpy as np
import time
import matplotlib.pyplot as plt


# Initialize variables for this exerise
x=np.linspace(-4,4,2000,endpoint=False)
y=np.linspace(-4,4,2000,endpoint=False)
fxy_non_vec=np.zeros((2000,2000))     #f(x,y) (non-vectorized)
fxy_thre_non_vec=np.zeros((2000,2000))  #thresholded f(x,y) (non-vectorized)
fxy_thre_vec=np.zeros((2000,2000))   #thresholded f(x,y) (vectorized)
x_squa=x*x
y_squa=y*y
x_squa,y_squa=np.meshgrid(x_squa,y_squa)

# Nonvectorized implementation
start_non_vec=time.time()
for i in range(0,2000):
    for j in range(0,2000):
        fxy_non_vec[j,i]=(-4+8/2000*i)**2-2*(-4+8/2000*j)**2
        if fxy_non_vec[j,i]>0:
            fxy_thre_non_vec[j,i]=1
end_non_vec=time.time()
time_non_vec=end_non_vec-start_non_vec   #calculate running time

# Vectorized implementation
start_vec=time.time()
fxy_vec=x_squa-2*y_squa
fxy_thre_vec[fxy_vec>0]=1
end_vec=time.time()
time_vec=end_vec-start_vec   #calculate running time

# Print the time for each and the speed increase
print('Time [sec] (non-vectorized):' ,round(time_non_vec,5))
print('Time [sec] (vectorized):    ' ,round(time_vec,5))
print('The vectorized code is {} times faster than the non-vectorized code'.format(round(time_non_vec/time_vec,5)))

# Plot the result
plt.figure()
plt.imshow(fxy_non_vec,extent=(-4,4,-4,4))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('f(x,y) (non-vecotrized)')
plt.show()

plt.figure()
plt.imshow(fxy_thre_non_vec,extent=(-4,4,-4,4))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('thresholded f(x,y) (non-vecotrized)')
plt.show()


plt.figure()
plt.imshow(fxy_vec,extent=(-4,4,-4,4))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('f(x,y) (vecotrized)')
plt.show()

plt.figure()
plt.imshow(fxy_thre_vec,extent=(-4,4,-4,4))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('thresholded f(x,y) (vecotrized)')
plt.show()
```

**ANSWER**

for some error terms $e_i$.
